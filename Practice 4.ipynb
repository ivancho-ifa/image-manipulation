```python
import cv2 as cv
import matplotlib.pyplot as plt
import numpy as np

colors = np.arange(0, 256)
figure_size = (40, 10)
histogram_bins_edges = range(0, 257)

circuit = cv.imread('resources/practice-4/circuit.png', cv.IMREAD_GRAYSCALE)
coins = cv.imread('resources/practice-4/coins.png', cv.IMREAD_GRAYSCALE)
gray_text = cv.imread('resources/practice-4/gray_text.png', cv.IMREAD_GRAYSCALE)

_, gray_text_binarized = cv.threshold(gray_text, 160, 255, cv.THRESH_BINARY)
_, coins_binarized = cv.threshold(coins, 100, 255, cv.THRESH_BINARY)
coins_binarized = cv.bitwise_not(coins_binarized)
```

```python
# 1. Приложете морфологическата операция ерозия със структурен елемент 2х2 върху бинаризацията на gray_text.gif.

kernel = np.ones((2, 2), np.uint8)
gray_text_eroded = cv.erode(gray_text_binarized, kernel)

plt.figure(dpi=300)
plt.imshow(np.concatenate((gray_text, gray_text_binarized, gray_text_eroded), axis=1), 'gray', vmin=0, vmax=255)
```

```python
# 2. Приложете морфологическата операция дилатация със структурен елемент 2х2 върху бинаризацията на gray_text.gif.

kernel = np.ones((2, 2), np.uint8)
gray_text_dilated = cv.dilate(gray_text_binarized, kernel)

plt.figure(dpi=300)
plt.imshow(np.concatenate((gray_text, gray_text_binarized, gray_text_dilated), axis=1), 'gray', vmin=0, vmax=255)
```

```python
# 3. Приложете морфологическата операция отваряне със структурен елемент 2х2 върху бинаризацията на gray_text.gif.

kernel = np.ones((2, 2), np.uint8)
gray_text_opening = cv.morphologyEx(gray_text_binarized, cv.MORPH_OPEN, kernel)

plt.figure(dpi=300)
plt.imshow(np.concatenate((gray_text, gray_text_binarized, gray_text_opening), axis=1), 'gray', vmin=0, vmax=255)
```

```python
# 4. Приложете морфологическа операция отваряне със структурен елемент 5х5 и 9x9 върху circuit.gif.

kernel_5 = np.ones((5, 5), np.uint8)
circuit_opening_5 = cv.morphologyEx(circuit, cv.MORPH_OPEN, kernel_5)

kernel_9 = np.ones((9, 9), np.uint8)
circuit_opening_9 = cv.morphologyEx(circuit, cv.MORPH_OPEN, kernel_9)

plt.figure(dpi=300)
plt.imshow(np.concatenate((circuit, circuit_opening_5, circuit_opening_9), axis=1), 'gray', vmin=0, vmax=255)
```

```python
# 5. Използвайте морфологически операции, за да подготвите изображението coins.gif за
# последваща обработка. Последващата обработка ще трябва да преброи монетите,
# чрез алгоритъм, който запълва области

kernel = np.ones((5, 5), np.uint8)
coins_eroded = cv.erode(coins_binarized, kernel, iterations=4)

plt.figure(dpi=300)
plt.imshow(np.concatenate((coins, coins_binarized, coins_eroded), axis=1), 'gray', vmin=0, vmax=255)
```
