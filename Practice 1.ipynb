```python
import numpy
import matplotlib.pyplot as plot

colors = numpy.arange(0, 256)
histogram_bins_edges = range(0, 257)

figure_size = (32, 4)
```

```python
# 1. Създайте вектор от 100 променливи със стойности между 0 и 255. Пресметнете и визуализирайте хистограмата на получените стойности.

# Create vector with 100 elements of [color.min(), colors.max()]
flat_image = numpy.random.randint(colors.min(), colors.max(), 100)
print(f'flat_image: {flat_image}, {len(flat_image)}')

# Calculate histogram
histogram, _ = numpy.histogram(flat_image, bins=histogram_bins_edges)
print(f'Histogram of flat_image: {histogram}, {len(histogram)}')

# Calculate histogram manually
histogram_manual = numpy.zeros(len(colors), dtype=int)
for pixel in flat_image:
    histogram_manual[pixel] += 1
print(f'Manually calculated histogram of flat_image: {histogram_manual}, {len(histogram_manual)}')

# Check if the histograms are correct
assert numpy.array_equal(histogram, histogram_manual)

# Visualize histogram
# plot.figure creates a new plotting picture
plot.figure(figsize=figure_size)

# Visualize histogram method #1
plot.plot(histogram)

# Visualize histogram method #2 no need to 
plot.hist(flat_image, bins=colors)
```

```python
# 2. Пресметнете честотното разпределение на стойностите от Зад. 1 и го визуализирайте.

# Calculate cummulative distribution funciton (честотно разпределение)
cdf = numpy.cumsum(histogram)

# Visualize CDF
print(f'CDF: {cdf}, {len(cdf)}')
plot.figure(figsize=figure_size)
plot.plot(colors, cdf)

# Normalize CDF to the histogram
cdf_normalized_to_histogram = cdf * float(histogram.max()) / cdf.max()

# Visualize CDF to the histogram
print(f'CDF normalized to histogram: {cdf_normalized_to_histogram}')
plot.figure(figsize=figure_size)
plot.hist(flat_image, bins=colors)
plot.plot(colors, cdf_normalized_to_histogram)
```

```python
# 3. Изравнете хистограмата на стойностите от Зад. 1.

import cv2

# Equalize the histogram using OpenCV
equalized_image_opencv = cv2.equalizeHist(flat_image.astype(numpy.uint8)).flatten()

# Visualize equalized image with OpenCV
print(f'flat_image            : {flat_image}')
print(f'equalized_image_opencv: {equalized_image_opencv}')

# Visualize equalized histogram
plot.figure(figsize=figure_size)
plot.hist(flat_image, bins=colors)
plot.hist(equalized_image_opencv, bins=colors)

# Equalize the histogram manually

print(f'cdf[115]: {cdf[115]}')

# Normalize CDF to the colors. This is the formula from the lectures used for equalization of histograms
min_positive = min(filter(lambda v: v > 0, cdf))
cdf_normalized_to_colors = numpy.around((cdf -  min_positive) * float(len(colors) - 1) / float(len(flat_image) - 1)).astype(int)
print(f'CDF normalized to the color space: {cdf_normalized_to_colors}')

# Visualize normalized to colors CDF
plot.figure(figsize=figure_size)
plot.plot(colors, cdf_normalized_to_colors)

# Equalize histogram of image

# A shortcut to do the code below
equalized_image = cdf_normalized_to_colors[flat_image]
# equalized_image = numpy.zeros(len(flat_image), dtype=int)
# for i, pixel in enumerate(flat_image):
#     equalized_image[i] = cdf_normalized_to_colors[pixel]
    
# Visualize equalized image
print(f'flat_image     : {flat_image}')
print(f'equalized_image: {equalized_image}')

# Visualize equalized histogram
plot.figure(figsize=figure_size)
plot.hist(flat_image, bins=colors)
plot.hist(equalized_image, bins=colors)

# Compare equalized histograms
print(f'flat_image            : {flat_image}')
print(f'equalized_image       : {equalized_image}')
print(f'equalized_image_opencv: {equalized_image_opencv}')
plot.figure(figsize=figure_size)
plot.hist(flat_image, bins=colors)
plot.hist(equalized_image, bins=colors)
plot.hist(equalized_image_opencv, bins=colors)
# OpenCV uses another algorithms and sometimes the values differ, so comparison of every element is not applicable
```

```python
# 4. Изравнете хистограмата на матрица с размерност 100х200 от случайни стойности между 0 и 255.

# Generate random matrix 100x200
image_size = (100, 200)
image = numpy.random.randint(colors.min(), colors.max(), image_size)
print(f'image: {image}, {len(image)}x{len(image[0])}')

# Equalize image using OpenCV
equalized_image_opencv = cv2.equalizeHist(image.astype(numpy.uint8))
print(f'equalized_image_opencv: {equalized_image_opencv}, {len(equalized_image_opencv)}x{len(equalized_image_opencv[0])}')

# Equalize image manually

flat_image = image.flatten()
histogram, _ = numpy.histogram(flat_image, bins=histogram_bins_edges)

cdf = numpy.cumsum(histogram)
min_positive = min(filter(lambda v: v > 0, cdf))
cdf_normalized_to_colors = numpy.around((cdf -  min_positive) * float(len(colors) - 1) / float(len(flat_image) - 1)).astype(int)

equalized_image = cdf_normalized_to_colors[flat_image]
equalized_image = numpy.reshape(equalized_image, image_size)
print(f'equalized_image: {equalized_image}, {len(equalized_image)}x{len(equalized_image[0])}')

plot.figure(figsize=figure_size)
plot.hist(flat_image, bins=colors)
plot.hist(equalized_image.flatten(), bins=colors)
plot.hist(equalized_image_opencv.flatten(), bins=colors)
```

```python
# 5. Преобразувайте изображението cube.jpg от цветно към тонове на сивото.

# The BGR color scheme is default
cube_bgr = cv2.imread('resources/practice-1/cube.jpg')
plot.figure()
plot.imshow(cube_bgr)

cube_grayscale = cv2.cvtColor(cube_bgr, cv2.COLOR_BGR2GRAY)
plot.figure()
# This is the correct way to show a grayscale image in matplotlib. Explanation in #7
plot.imshow(cube_grayscale, 'gray', vmin=0, vmax=255)
```

```python
# 7. Изравнете хистограмата на gray.jpg.

gray = cv2.imread('resources/practice-1/gray.jpg', cv2.IMREAD_GRAYSCALE)
plot.figure()
# Wrong way to show grayscale images
plot.imshow(gray, 'gray')
plot.figure()
# We need to specify the data range so that matplotlib doesn't guess the data range based on the data
# Otherwise the lowest value pixel becomes black (0) and the highest - white (255). 
plot.imshow(gray, 'gray', vmin=0, vmax=255)

gray_equalized = cv2.equalizeHist(gray.astype(numpy.uint8))
plot.figure()
plot.imshow(gray_equalized, 'gray', vmin=0, vmax=255)
```

```python
# 8. Изравнете хистограмите на R, G и B компонентите на color.jpg.

color_bgr = cv2.imread('resources/practice-1/color.jpg')
plot.figure()
plot.imshow(color_bgr)

# To show the true colors of an image using plot you need to convert it to RGB
color_rgb = cv2.cvtColor(color_bgr, cv2.COLOR_BGR2RGB)
plot.figure()
plot.imshow(color_rgb)

blue, green, red = cv2.split(color_rgb)
red = cv2.equalizeHist(red)
green = cv2.equalizeHist(green)
blue = cv2.equalizeHist(blue)

equalized = cv2.merge((blue, green, red))
plot.figure()
plot.imshow(equalized)
```

```python
# 9. Изравнете хистограмата на компонентата на яркостта на color.jpg.

# The V component is brightness
color_hsv = cv2.cvtColor(color_bgr, cv2.COLOR_BGR2HSV)
hue, saturation, value = cv2.split(color_hsv)

value = cv2.equalizeHist(value)
equalized_brightness = cv2.merge((hue, saturation, value))

equalized_brightness = cv2.cvtColor(equalized_brightness, cv2.COLOR_HSV2RGB)
plot.figure()
plot.imshow(equalized_brightness)

```
