```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

### Задача 1.

Приложете морфологическата операция ерозия със структурен елемент 2х2 върху
бинаризацията на gray_text.gif.

```python
input = cv2.imread('gray_text.png', cv2.IMREAD_GRAYSCALE)

bin = np.copy(input)
t = 160
bin[bin < t] = 0
bin[bin >= t] = 255

kernel = np.ones((2, 2), np.uint8)

erode = cv2.erode(bin, kernel)

plt.figure(figsize=(40, 10))
plt.imshow(np.concatenate((bin, erode), axis=1), 'gray')
```

### Задача 2.

Приложете морфологическата операция дилатация със структурен елемент 2х2 върху
бинаризацията на gray_text.gif.

```python
input = cv2.imread('gray_text.png', cv2.IMREAD_GRAYSCALE)

bin = np.copy(input)
t = 160
bin[bin < t] = 0
bin[bin >= t] = 255

kernel = np.ones((2, 2), np.uint8)

dilated = cv2.dilate(bin, kernel)

plt.figure(figsize=(40, 10))
plt.imshow(np.concatenate((bin, dilated), axis=1), 'gray')
```

### Задача 3.

Приложете морфологическата операция отваряне със структурен елемент 2х2 върху
бинаризацията на gray_text.gif.

```python
input = cv2.imread('gray_text.png', cv2.IMREAD_GRAYSCALE)

bin = np.copy(input)
t = 160
bin[bin < t] = 0
bin[bin >= t] = 255

kernel = np.ones((2, 2), np.uint8)

open = cv2.morphologyEx(bin, cv2.MORPH_OPEN, kernel)

plt.figure(figsize=(40, 10))
plt.imshow(np.concatenate((bin, open), axis=1), 'gray')
```

### Задача 4.

Приложете морфологическа операция отваряне със структурен елемент 5х5 и 9x9
върху circuit.gif.

```python
input = cv2.imread('circuit.png', cv2.IMREAD_GRAYSCALE)

kernel_5 = np.ones((5, 5), np.uint8)
open_5 = cv2.morphologyEx(input, cv2.MORPH_OPEN, kernel_5)

kernel_9 = np.ones((9, 9), np.uint8)
open_9 = cv2.morphologyEx(input, cv2.MORPH_OPEN, kernel_9)

plt.figure(figsize=(40, 10))
plt.imshow(np.concatenate((input, open_5, open_9), axis=1), 'gray')
```
